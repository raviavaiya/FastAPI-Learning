```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
carbs = float(input("Enter Carbohydrates (g): "))
protein = float(input("Enter Protein (g): "))
fat = float(input("Enter Fats (g): "))

X = df[["Carbohydrates (g)", "Protein (g)", "Fats (g)"]].values
target = [[carbs, protein, fat]]

# Cosine similarity
similarity_scores = cosine_similarity(target, X)[0]

# Get top 5 indices sorted by similarity
top_5_indices = similarity_scores.argsort()[-5:][::-1]

# Print results
print("\n✅ Top 5 Most Similar Dishes (Cosine Similarity):\n")
for idx in top_5_indices:
    dish = df.iloc[idx]
    score = similarity_scores[idx] * 100
    print(f"{dish['Dish Name']} — Similarity: {score:.2f}%")
```
